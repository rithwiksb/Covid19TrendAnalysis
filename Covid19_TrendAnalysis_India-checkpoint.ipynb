```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/imdevskp/covid-19-india-data/refs/heads/master/complete.csv")
```

```python
confirmed_cases = data['Total Confirmed cases']
```

```python
cured = data['Cured/Discharged/Migrated']
```

```python
plt.title("Confirmed Cases vs Cured Cases Scatter Plot")
plt.scatter(confirmed_cases,cured,s=5)
plt.xlabel("Confirmed Covid-19 Cases")
plt.ylabel("Cured Cases of Covid-19")
plt.show()
```

# Number of cases vs deaths in Telangana using Line Plot

```python
telangana_data = data[data['Name of State / UT']=="Telangana"]
```

```python
telangana_data
```

```python
plt.title("Number of Cases vs Number of Recoveries in Telangana")
plt.plot(telangana_data['Total Confirmed cases'],telangana_data['Cured/Discharged/Migrated'], label='Deaths', color='red', linestyle='-')
plt.xlabel("Number of Cases")
plt.ylabel("Number of Recoveries")
plt.show()
```
